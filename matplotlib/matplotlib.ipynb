# import library

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# 线图 line graph

```python
X = np.linspace(-10, 10, 100)
Y = 1.0 / (1.0 + np.power(np.e, -1 * X))
#plt.plot(X, Y, 'r--')
plt.plot(X, Y, color='r', linestyle='--', label='sigmoid')
plt.xlim(-10, 10)
plt.xlabel("x")
plt.ylim(0, 1)
plt.ylabel("y")
plt.title('scatter')
plt.legend(loc='upper left')
plt.show()
```

# 散点图 scatter

```python
X = np.random.randint(0, 100, 50)
Y = np.random.randint(0, 100, 50)
plt.scatter(X, Y)
plt.show()
```

# 直方图 histogram

```python
X = np.random.randn(1000)
Y = np.random.randn(1000) + 5
plt.hist(X, bins=30)
plt.show()
plt.hist2d(X, Y, bins=30)
plt.show()
```

# 数字标签 注释 移动轴

```python
X = np.linspace(0, 10, 100)
Y = np.sin(X) * 10

plt.figure(figsize=(10, 6))
# 添加正弦函数图
plt.plot(X, Y, 'r-')
# 添加条形图
plt.bar(-5, 8, width=0.3, color='c')
# 设置数字标签
plt.text(-5, 8+0.2, '8', ha='center', va= 'bottom',fontsize=10)
# 设置spine为(0,0)
ax = plt.gca()
ax.spines['right'].set_color('None')
ax.spines['top'].set_color('None')
ax.spines['left'].set_position(('data',0))
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
# 添加虚线
plt.plot([7, 7], [0, np.sin(7)*10], 'b--')
# 添加点
plt.scatter([7], [np.sin(7)*10], 50, color='r')
# 添加注释
plt.annotate('annotate', xy=[7, np.sin(7)*10], xycoords='data',  
             xytext=(-80, 30), textcoords='offset points',
             fontsize=16,arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.show()
```

# 子图 subplot

```python
fig = plt.figure(figsize=(11, 6))

for i, title in enumerate(['sin func', 'cos func']):
    ax = fig.add_subplot(1,2,i+1)
    X = np.linspace(-2*np.pi, 2*np.pi, 100)
    if title == 'sin func':
        Y = np.sin(X) 
    else:
        Y = np.cos(X) 
    ax.plot(X, Y, 'r-')
    ax.set_title(title)
    
fig.tight_layout()
fig.show()
```
